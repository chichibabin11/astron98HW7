```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('planets')

#scatter plot of orbital period on x-axis and mass on y-axis (use log scale)
#individual points coloured by discovery method
sns.scatterplot(x = 'orbital_period', y = 'mass', hue = 'method', data = df, palette = 'Set1', alpha = 0.7)
plt.xscale('log')
plt.xlabel('Orbital Period')
plt.ylabel('Mass')
plt.title('Orbital Period vs. Mass')
plt.legend(title = 'Discovery Method', bbox_to_anchor = (1,1))
plt.grid(True)
plt.show()
```

```python
#bar chart: how many exoplanets discovered by year
#categorized with different colours accoding to discovery methods
#remove NaNs

df = sns.load_dataset('planets')
df = df.dropna(subset = ['orbital_period', 'mass', 'method'])

discovery_by_year = df.groupby(['year', 'method']).size().unstack(fill_value = 0)
discovery_by_year.plot(kind = 'bar', stacked = True, colormap = 'viridis')
plt.xlabel('Year')
plt.ylabel('Number of Exoplanets')
plt.title('Number of Exoplanets Discovered per Year and Methods')
plt.legend(title = 'Discovery Method')
plt.grid(axis = 'y')
plt.show()
```
