```python
#tweet analysis
import pandas as pd
```

```python
df = pd.read_json('AOC_recent_tweets.txt', orient = 'records')
df.set_index('id', inplace = True)
df
```

```python
#function: hour + minute/60 + seconds/60^2
def time_in_hours(datetime_col):
    dec_hours = datetime_col.dt.hour + datetime_col.dt.minute/60 + datetime_col.dt.second/3600
    return dec_hours

df['created_at'] = pd.to_datetime(df['created_at'])
df['hours'] = time_in_hours(df['created_at'])
df_save = df[['created_at', 'hours', 'full_text']]
df_save.to_csv('AOC_tweets.csv')
```
